## 4. Simple Linear Regression:

##### ⁍  Linear regression is used to predict the value of an outcome variable Y based on one or more input predictor variables X.

```r
model<- lm(Temperature ~ Weekly_Sales, data = m3)
```

```r
model
```

```r
ggplot(model.diag.metrics, aes(Temperature, Weekly_Sales)) +
  geom_point() +
  stat_smooth(method = lm, se = FALSE) +
  geom_segment(aes(xend = Temperature, yend = .fitted), color = "red", size = 0.3)
```

# ❔How well can temperatures predict the weekly sales

✅ Tempreature being statiscally non-sinificant as p value is not < 0.05 in all the comparisons, we can state that Temperature ~ 
Weekly_Sales cannot reject NULL hypothesis. We can see from the above graph there are many residual errors too, hence we can state that Temperature is not a good choice in predicting the weekly sales.

```r
par(mfrow = c(2, 2))
plot(model)
```

```r
summary(model)
```

##### ⁍ From the summary(model) we can state that  Temperature~Weekly_Sales is not statistically significant as the p value is < 0.05 hence we can safely reject the NULL hypothesis. t value is -1.501 which indicates that it is more likely that the coefficient is equal to zero purely by chance. The Adj R-Squared gives us the total variance and that is 2.975e-06. F-statistics measures the goodness of fit which is 2.254.
